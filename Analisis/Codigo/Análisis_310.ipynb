## **Test de hipótesis**

```python
from statsmodels.formula.api import ols
from scipy.stats import shapiro
from scipy.stats import kruskal
from scipy.stats import mannwhitneyu
```

```python
model = ols('Ordenes ~ C(Promocion_cod) + C(Tipo_promocion_cod) + C(Feriado_cod) + C(Mes_cod) + C(Dia_semana_cod) + C(Numero_semana) + C(Funcionamiento_cod) + C(Estacion_cod) + C(Fin_de_semana_cod)', data=df_310_dia).fit()
```

```python
residuals = model.resid
shapiro_test = shapiro(residuals)
print(f'Shapiro-Wilk test: W={shapiro_test[0]}, p-value={shapiro_test[1]}')
```

---
Como el p-valor es menor que 0,05 se rechaza la hipótesis nula, por lo que los residuos no están normalmente distribuidos.
No se puede realizar ANOVA, por lo que se implementará su versión no paramétrica: **Kruskal Wallis**

**Tipo de promoción**

```python
ordenes_tipo_promo = [group['Ordenes'].tolist() for name, group in df_310_dia.groupby('Tipo_promocion_cod')]
kruskal_test = kruskal(*ordenes_tipo_promo)
print(f'Kruskal-Wallis test: H={kruskal_test[0]}, p-value={kruskal_test[1]}')
```

Si hay diferencia significativa, se aplica post hoc para ver que promociones tienen mas peso

```python
df_promocion = [df_310_dia[df_310_dia['Tipo_promocion_cod'] == "0"]['Ordenes'],
                df_310_dia[df_310_dia['Tipo_promocion_cod'] == "1"]['Ordenes'],
                df_310_dia[df_310_dia['Tipo_promocion_cod'] == "2"]['Ordenes'],
                df_310_dia[df_310_dia['Tipo_promocion_cod'] == "3"]['Ordenes'],
                df_310_dia[df_310_dia['Tipo_promocion_cod'] == "4"]['Ordenes'],
                df_310_dia[df_310_dia['Tipo_promocion_cod'] == "5"]['Ordenes'],
                df_310_dia[df_310_dia['Tipo_promocion_cod'] == "6"]['Ordenes'],
                df_310_dia[df_310_dia['Tipo_promocion_cod'] == "7"]['Ordenes'],
                df_310_dia[df_310_dia['Tipo_promocion_cod'] == "8"]['Ordenes'],
                df_310_dia[df_310_dia['Tipo_promocion_cod'] == "9"]['Ordenes'],
                df_310_dia[df_310_dia['Tipo_promocion_cod'] == "10"]['Ordenes'],
                df_310_dia[df_310_dia['Tipo_promocion_cod'] == "11"]['Ordenes']]
```

```python
dunn_test = sp.posthoc_dunn(df_promocion)
```

```python
group_names= ["Sin promoción", "Black","Blackout","Cyber","Días T","T week","Post black","Post Cyber","Promo año nuevo","Promo fiestas patrias","Promo navidad","Venta nocturna"]
dunn_test.columns= group_names
dunn_test.index= group_names
dunn_test.style.applymap(lambda x: "background-color:red" if x<0.05 else "background-color: white")
```

Se observa que las promociones que presentan distinta distribución en las ordenes, en comparación con la media sin promoción son las siguientes:
- Blackout
- Cyber
- Días T
- T week
- Post cyber
- Promo año nuevo
- Promo fiestas patrias
- Promo navidad
- Venta nocturna

---
Para variables categóricas donde se deben comparar dos grupos, se usa la prueba de **Mann Whitney**, versión no parámetrica de t student

**Promocion**

```python
ordenes = df_310_dia.groupby('Promocion_cod')['Ordenes'].apply(list).tolist()
ordenes_sin_promo, ordenes_promo = ordenes
```

```python
mann_whitney_test = mannwhitneyu(ordenes_sin_promo, ordenes_promo, alternative='two-sided')
print(f'Mann-Whitney U test: U={mann_whitney_test.statistic}, p-value={mann_whitney_test.pvalue}')
```

Si hay diferencia significativa en las ordenes con la existencia de promoción

**Feriado**

```python
ordenes = df_310_dia.groupby('Feriado_cod')['Ordenes'].apply(list).tolist()
ordenes_no_feriado, ordenes_feriado = ordenes
```

```python
mann_whitney_test = mannwhitneyu(ordenes_no_feriado, ordenes_feriado, alternative='two-sided')
print(f'Mann-Whitney U test: U={mann_whitney_test.statistic}, p-value={mann_whitney_test.pvalue}')
```

Si hay diferencia significativa, no es muy notoria
