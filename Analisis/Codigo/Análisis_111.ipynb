## **Test de hipótesis**

```python
from statsmodels.formula.api import ols
from scipy.stats import shapiro
from scipy.stats import kruskal
from scipy.stats import mannwhitneyu
```

```python
model = ols('Ordenes ~ C(Promocion_cod) + C(Tipo_promocion_cod) + C(Feriado_cod) + C(Mes_cod) + C(Dia_semana_cod) + C(Numero_semana) + C(Funcionamiento_cod) + C(Estacion_cod) + C(Fin_de_semana_cod)', data=df_111_dia).fit()
```

```python
residuals = model.resid
shapiro_test = shapiro(residuals)
print(f'Shapiro-Wilk test: W={shapiro_test[0]}, p-value={shapiro_test[1]}')
```

---
Como el p-valor es menor que 0,05 se rechaza la hipótesis nula, por lo que los residuos no están normalmente distribuidos.
No se puede realizar ANOVA, por lo que se implementará su versión no paramétrica: **Kruskal Wallis**

**Tipo de promoción**

```python
df_promocion = [df_111_dia[df_111_dia['Tipo_promocion_cod'] == "0"]['Ordenes'],
                df_111_dia[df_111_dia['Tipo_promocion_cod'] == "1"]['Ordenes'],
                df_111_dia[df_111_dia['Tipo_promocion_cod'] == "2"]['Ordenes'],
                df_111_dia[df_111_dia['Tipo_promocion_cod'] == "3"]['Ordenes'],
                df_111_dia[df_111_dia['Tipo_promocion_cod'] == "4"]['Ordenes'],
                df_111_dia[df_111_dia['Tipo_promocion_cod'] == "5"]['Ordenes'],
                df_111_dia[df_111_dia['Tipo_promocion_cod'] == "6"]['Ordenes'],
                df_111_dia[df_111_dia['Tipo_promocion_cod'] == "7"]['Ordenes'],
                df_111_dia[df_111_dia['Tipo_promocion_cod'] == "8"]['Ordenes'],
                df_111_dia[df_111_dia['Tipo_promocion_cod'] == "9"]['Ordenes'],
                df_111_dia[df_111_dia['Tipo_promocion_cod'] == "10"]['Ordenes'],
                df_111_dia[df_111_dia['Tipo_promocion_cod'] == "11"]['Ordenes']]
```

```python
dunn_test = sp.posthoc_dunn(df_promocion)
```

```python
group_names= ["Sin promoción", "Black","Blackout","Cyber","Días T","T week","Post black","Post Cyber","Promo año nuevo","Promo fiestas patrias","Promo navidad","Venta nocturna"]
dunn_test.columns= group_names
dunn_test.index= group_names
dunn_test.style.applymap(lambda x: "background-color:red" if x<0.05 else "background-color: white")
```

Se observa que las promociones que presentan distinta distribución en las ordenes, en comparación con la media sin promoción son las siguientes:
- Cyber
- T week
- Promo fiestas patrias
- Venta nocturna

Para variables categóricas donde se deben comparar dos grupos, se usa la prueba de **Mann Whitney**, versión no parámetrica de t student

**Promocion**

```python
ordenes = df_111_dia.groupby('Promocion_cod')['Ordenes'].apply(list).tolist()
ordenes_sin_promo, ordenes_promo = ordenes
```

```python
mann_whitney_test = mannwhitneyu(ordenes_sin_promo, ordenes_promo, alternative='two-sided')
print(f'Mann-Whitney U test: U={mann_whitney_test.statistic}, p-value={mann_whitney_test.pvalue}')
```

No es significativa

**Feriado**
